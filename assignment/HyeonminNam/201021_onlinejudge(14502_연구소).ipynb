```python
# 14502
## <문제 풀이 알고리즘> - BFS, 완전탐색 활용

# 1. 지도 정보 바탕으로 안전 영역 개수 세기
# 지도 정보 바탕으로 안전 영역 개수 측정할 수 있는 함수 작성
# 바이러스 시작점 queue에 추가하고 BFS를 통해서 이웃들 탐색(이때마다 바이러스 개수 +1)
# 빈 칸이면 바이러스 전염시키고 queue에 추가시켜서 다시 바이러스 시작점으로 삼음(이때마다 바이러스 개수 +1)
# queue에 추가되는 것이 없을 때까지 위 과정 반복
# 최종 안전 영역 개수(= 칸 개수(n*m) - 벽 개수 - 바이러스 개수) return

# 2. 벽 3개 놓았을 때 안전 영역 최댓값 
# combinations 활용하여 벽 3개 놓을 수 있는 모든 조합 탐색
# 벽 3개의 좌표 정보 지도에 추가한 뒤에 1번 과정 실시
# return된 값들 리스트에 추가
# max 취해서 출력


## 문제 풀이 코드
from collections import deque
from itertools import combinations
# import sys
# input = sys.stdin.readline
import copy

n, m = list(map(int, input().split()))

# 지도 정보 입력
org_map = []
blank = []
for x in range(n):
    line = list(map(int, input().split()))
    org_map.append(line)
    for y, num in enumerate(line):
        if num == 0:
            blank.append((x, y))


# 이동 방향(동, 서, 남, 북)
dx = [0, 0, 1, -1]
dy = [1, -1, 0, 0]

# 안전 영역 탐색 함수
def safe_zone(n, m, map_info, new_walls):
    # 새로운 벽 정보 업데이트
    for x, y in new_walls:
        map_info[x][y] = 1
    # 빈 칸, 벽, 바이러스 좌표 정리
    blank, wall, virus = [], [], []
    for x, line in enumerate(map_info):
        for y, num in enumerate(line):
            if num == 0:
                blank.append([x, y])
            elif num == 1:
                wall.append([x, y])
            else:
                virus.append([x, y])
    # 바이러스 개수 세기
    virus_num = 0
    for v in virus:
        # queue에는 바이러스의 좌표들이 들어감
        queue = deque([v])
        while queue:
            center = queue.popleft()
            virus_num += 1
            # 동서남북 이웃 판단
            for i in range(4):
                # 이웃 좌표
                neighbor = center[0]+dx[i], center[1]+dy[i]
                n_x, n_y = neighbor[0], neighbor[1]
                # 이동해서 지도를 벗어나지 않았고,
                if 0 <= n_x < n and 0 <= n_y < m:
                    # 이웃이 빈 칸일 경우 바이러스 전염시키고, queue에 추가
                    if map_info[n_x][n_y] == 0:
                        map_info[n_x][n_y] = 2
                        queue.append([n_x, n_y])
    safe_num = n*m - len(wall) - virus_num
    return safe_num

# 벽 3개 세우는 경우의 수(조합) 
safe_num_lst = []
new_wall_comb = combinations(blank, 3)
for new_walls in list(new_wall_comb):
    new_map = copy.deepcopy(org_map)
    safe_num_lst.append(safe_zone(n, m, new_map, new_walls))
print(max(safe_num_lst))
```
