뱀 문제

```python
from collections import deque
import numpy as np

N = int(input())
# N*N행렬의 바깥부분을 1로 패딩하기위해 (N+2) * (N+2)행렬로 초기화
ls2 = [[0] * (N+2) for _ in range(N+2) ]
for i in range(N+2) :
    for j in range(N+2) :
        if i == 0 or i == N+1 :
            ls2[i][j] = 1
        elif j == 0 or j == N+1 :
            ls2[i][j] = 1


K = int(input())
for _ in range(K) :
    a, b = map(int, input().split())
    #행렬에 입력받은 사과 2로 배치
    ls2[a][b] = 2

L = int(input())

snake_control_array = []

for _ in range(L) :
    X, C = map(str, input().split())
    snake_control_array.append([int(X), C])

ls2[1][1] ='@'
snake_len = 1
snake_loc_ls = deque([])
snake_loc_ls.append([1,1]) #시작위치 삽입
direction = 0 # 초기 방향은 동쪽
time = 0 
condition = False

dx = [1, 0, -1, 0]
dy = [0, 1, 0, -1]

def course(way, C) :
    if C == "D" : 
        way = (way +1) % 4
    else :
        way = (way -1) % 4
    return way
    

for s in snake_control_array :
    #강제 종료 조건
    if condition :      
        break
    for _ in range(s[0]- time) :
        print('{}초'.format(time))
        print(np.array(ls2))
        time += 1
        last_loc = snake_loc_ls[-1]
        #뱀의 이동 위치 추가
        snake_loc_ls.append([last_loc[0] + dy[direction],last_loc[1] + dx[direction]])
        
        #이동경로에 사과가 있다면 뱀의 길이 +1
        if object == 2 :
            snake_len += 1
            
        #뱀의 길이만큼 뱀의 꼬리 위치 조정
        if snake_len < len(snake_loc_ls) :
            ls2[snake_loc_ls[0][0]][snake_loc_ls[0][1]] = 0
            snake_loc_ls.popleft()
        
        #이동경로
        object = ls2[snake_loc_ls[-1][0]][snake_loc_ls[-1][1]]
        
        #이동경로가 벽이나 자신의 몸이라면 종료   
        if object == 1 or object == '@':
            condition = True
            break
        #아니라면 이동 경로에 뱀의 머리 위치
        else :
            ls2[snake_loc_ls[-1][0]][snake_loc_ls[-1][1]] = '@'
    
    #방향조정    
    direction = course(direction, s[1])
            
                
print(time)
```
