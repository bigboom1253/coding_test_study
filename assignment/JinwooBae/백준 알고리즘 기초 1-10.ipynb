###문제 : 알람시계

```
#원래 설정되어 있는 알람을 45분 앞서는 시간으로 바꾸는 것이다
#첫째 줄에 두 정수 H와 M이 주어진다. (0 ≤ H ≤ 23, 0 ≤ M ≤ 59) 그리고 이것은 현재 상근이가 설정한 놓은 알람 시간 H시 M분을 의미한다.
#입력 시간은 24시간 표현을 사용한다. 24시간 표현에서 하루의 시작은 0:0(자정)이고, 끝은 23:59(다음날 자정 1분 전)이다. 시간을 나타낼 때, 불필요한 0은 사용하지 않는다.
H, M = int(input()), int(input())
if M < 45 :
    if H < 1 :
        H += 23
        M += 60 - 45
    else :
        H -= 1
        M += 60 - 45
elif M >= 45 :
    M -= 45

print('{} {}'.format(H, M))
```

### 문제 : 더하기 사이클
(틀렸다고 뜸.. 어째서?)

```
# 원래의 숫자로 돌아올 때 까지 반복, 카운트
count = 0
N = int(input())
real_N = N
while 1 :
    namugi = int(N%10)
    if N/10 == 0 :
        new_number = int(namugi *10 + N)
    else :
        new_number = int(namugi * 10 + (N/10 + namugi)%10)
        count += 1
    
    if new_number == real_N :
        print(count)
        break
    else :
        N = new_number
```

###문제 : 평균은 넘겠지
(런타임 에러)

```
import numpy as np
test_case = int(input())

print_list =[]
for test in range(test_case) :
    case_ls = list(map(int, input().split())) #case_ls[0]은 학생 수, 이후는 점수
    #평균을 넘는 학생들의 비율을 반올림하여 소수점 셋째 자리까지 출력
    case_mean = np.mean(case_ls[1:]) #평균
    temp_list =list(np.sort(case_ls[1:]))
    for idx, score in enumerate(temp_list) :
        if score > case_mean :
            print_list.append(round(((case_ls[0] - idx)/case_ls[0])*100,3))
            break
for x in print_list :
    print('{}%'.format(x))
```

```
##-------런타임에러 발생이유---------
# 배열에 할당된 크기를 넘어서 접근했을 때
# 전역 배열의 크기가 메모리 제한을 초과할 때
# 지역 배열의 크기가 스택 크기 제한을 넘어갈 때
# 0으로 나눌 떄
# 라이브러리에서 예외를 발생시켰을 때
# 재귀 호출이 너무 깊어질 때
# 이미 해제된 메모리를 또 참조할 때
```

###문제 : 한수
(런타임 에러)

```
def hansu(N) :
    str_N = str(N)
    str_N_len = len(str_N)
    if str_N_len == 1 :
        count = 0
    elif str_N_len == 2 :
        count = 9
    elif str_N_len == 3 :
        count = 99

    for number in range(10**(str_N_len-1), N+1) :
        str_number = str(number)
        if len(str_number) == 3 :
            if int(str_number[0]) - int(str_number[1]) == int(str_number[1]) - int(str_number[2]) :
                count += 1 

        elif len(str_number) == 2 : 
            # 10~99는 등차수열로 간주
                count += 1   

        elif len(str_number) == 1 : 
            count += 1 # 1~9는 등차수열로 간주
            
        #1000은 등차수열이 아니므로 제외
    print(count)
N = int(input())
hansu(N)
```

```
#아스키코드를 이용 97부터 122까지 -97

S = input()
#알파벳 리스트를 생성, 인덱스를 참조하여 들어가자
# alphabet_ls = [chr(alphabet) for alphabet in range(97,123)]
alphabet_idx_ls = [-1 for x in range(26)]

for idx, alphabet in enumerate(S) :
    if alphabet_idx_ls[ord(alphabet) -97] == -1 :
        alphabet_idx_ls[ord(alphabet) -97] = idx

for x in alphabet_idx_ls :
    print(x, end =' ')
```

###문제 : 소수 찾기

```
N_count = int(input())
N_ls = list(map(int, input().split()))

#1. 1~1000 사이의 소수 리스트를 만들고 비교하며 카운트. #여러번 사용에 적합
#2. 소수 판별 함수 만들고, 비교 #적은 사용에 적합

def prime(number) :
    if number != 1:
        for x in range(2, number) :
            if number%x == 0 :
                return False
    else : 
        return False
    return True

count = 0
for i in N_ls :
    if prime(i) :
        count += 1

print(count)

```

###문제 : 직각삼각형

```
judge_ls = []
while True :
    triangle_line_ls = list(map(int, input().split()))
    if triangle_line_ls[0] == 0 :
        break
    triangle_line_ls.sort()
    if triangle_line_ls[0]**2 + triangle_line_ls[1]**2 == triangle_line_ls[2]**2 :
        judge_ls.append('right')
    else :
        judge_ls.append('wrong')
for x in judge_ls :
    print(x)
```
